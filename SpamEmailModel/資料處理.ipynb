```python
import pandas as pd 

df = pd.read_csv('test.csv')
df.head()
```

```python
df['sub_mssg'] = df['subject'] + ' ' + df['message']
df['sub_mssg'].head()
```

```python
df.drop('subject', axis=1, inplace=True)
df.drop('message', axis=1, inplace=True)
```

```python
df.head(5)
```

```python
import re
```

```python
import re
df['sub_mssg'] = df['sub_mssg'].apply(lambda x: re.sub(r'http\S+', '', x))
```

```python
#移除非中文、英文、数字的字符
# 以及多余的空格
def clean_text(text):
    if not isinstance(text, str):
        text = str(text)
    text = re.sub('[^\u4e00-\u9fa5a-zA-Z0-9]', ' ', text)
    text = re.sub('\s+', ' ', text).strip()
    return text
```

```python
cleaned = clean_text(df['sub_mssg'][1])

# 輸出結果
print("清理前：", df['sub_mssg'][1])
print("清理後：", cleaned)
```

```python
print(df['sub_mssg'][1])
```

```python
df['sub_mssg'] = df['sub_mssg'].apply(clean_text)
```

```python
print(df['sub_mssg'][1])
```

```python
import jieba
```

```python
#分詞化
def textParse(text):
    listOfTokens=jieba.lcut(text)
    return listOfTokens
```

```python
parsed = textParse(df['sub_mssg'][1])

# 輸出結果
print("分詞前：", df['sub_mssg'][1])
print("分詞後：", parsed)
```

```python
df['sub_mssg'] = df['sub_mssg'].apply(textParse)
```

```python
print(df['sub_mssg'][1])
```

```python
#載入停用詞表
def load_stopwords(filepath):
    with open(filepath, encoding='utf-8') as f:
        return set(line.strip() for line in f if line.strip())

stopword_list = load_stopwords('停用詞-繁體中文.txt')
```

```python
#刪除停用詞
def remove_stopwords(tokens):
    filtered_tokens = [token for token in tokens if token not in stopword_list]
    filtered_text = ' '.join(filtered_tokens)
    return filtered_text
```

```python
remove_stoped = remove_stopwords(df['sub_mssg'][1])

# 輸出結果
print("刪除前：", df['sub_mssg'][1])
print("刪除後：", remove_stoped)
```

```python
df['sub_mssg'] = df['sub_mssg'].apply(remove_stopwords)
```

```python
print(df['sub_mssg'][3])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

def tfidf_vectorize(texts):
    vectorizer = TfidfVectorizer(max_features=10000,
                                 max_df=0.85,
                                 min_df=3,
                                 norm='l2',
                                 smooth_idf=True,
                                 use_idf=True,
                                 ngram_range=(1,1))
    X = vectorizer.fit_transform(texts)
    return X, vectorizer
```

```python
X, vectorizer= tfidf_vectorize(df['sub_mssg'])
print(X.shape)
```

```python
print(X[1])
```

```python
# 取得詞彙表（注意：fit_transform 必須先執行過，vectorizer 已建立好）
feature_names = vectorizer.get_feature_names_out()

# 指定要查看第幾筆資料（第幾筆訊息）
row_index = 1
row = X[row_index]

# 把稀疏向量轉成欄位 index ➝ TF-IDF 值的 dict
sparse_dict = dict(zip(row.indices, row.data))

# 印出 index、對應的詞與該詞在該文件中的 TF-IDF 值
for idx in sorted(sparse_dict):
    word = feature_names[idx]
    tfidf = sparse_dict[idx]
    print(f"Index: {idx:5d}, Word: {word:20s}, TF-IDF: {tfidf:.5f}")
```

```python
import pandas as pd

# 指定想要看的文本索引
row_index = 1
row = X[row_index]

# 取得詞彙名稱（字串）
feature_names = vectorizer.get_feature_names_out()

# 轉成對照 dict：index ➝ tfidf
sparse_dict = dict(zip(row.indices, row.data))

# 建立 DataFrame
df_tfidf = pd.DataFrame({
    'Index': list(sparse_dict.keys()),
    'Word': [feature_names[idx] for idx in sparse_dict.keys()],
    'TF-IDF': list(sparse_dict.values())
})

# 排序（由大到小）
df_tfidf = df_tfidf.sort_values(by='TF-IDF', ascending=False)

# 顯示前幾筆（或不加就是全部顯示）
display(df_tfidf)
```

```python
import pandas as pd

# 取得詞彙表
feature_names = vectorizer.get_feature_names_out()

# 建立儲存所有結果的列表
rows = []

# 遍歷每一筆文本
for row_index in range(X.shape[0]):
    row = X[row_index]
    sparse_dict = dict(zip(row.indices, row.data))
    
    for idx in sparse_dict:
        word = feature_names[idx]
        tfidf = sparse_dict[idx]
        rows.append((row_index, word, tfidf))

# 建立 DataFrame
df_all = pd.DataFrame(rows, columns=['row_index', 'word', 'tfidf'])

# 儲存為 CSV
df_all.to_csv("all_tfidf_values.csv", index=False, encoding='utf-8-sig')
```
